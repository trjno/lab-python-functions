def initialize_inventory(products):

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory
````

````python
def total_price_orders(customer_orders):
    total_price = 0
    for element in customer_orders:
        is_valid = False
        while not is_valid:
            try:
                price = int(input("Input the price for the "+element))
                if price >= 0:
                    total_price += price
                    is_valid = True
                else:
                    print("Price can\'t be negative. Please Enter a valid price")
            except ValueError:
                print("Invalid input. Please enter a valid price")

    return total_price
````

````python
def get_customer_orders(inventory):

    customer_orders = set()
    
    is_valid_order_num = False
    while not is_valid_order_num:
        try:
            total_unique_product = int(input("How many product\'s type you will order ? "))
            if total_unique_product < 0:
                print("The number of product can\t be negative or null.")
            else:
                is_valid_order_num = True
        except ValueError:
            print("Invalid input. Please enter a valid order number.")
    for i in range(total_unique_product):
        is_valid_name = False
        while not is_valid_name:
            try:
                product_name = input("Input the  name of the product no "+str(i+1))
                if inventory[product_name] > 0:
                    customer_orders.add(product_name)
                    is_valid_name = True
                else:
                    print("This product is not available. Please, enter a valid product name")
            except:
                print("Input a valid product name")
    return customer_orders
````

````python
inventory = initialize_inventory(products)
customer_orders = get_customer_orders(inventory)

print(total_price_orders(customer_orders))
````
